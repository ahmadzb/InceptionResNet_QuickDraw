```python
import matplotlib.pyplot as plt
import pickle
```

Creating some lists which tell us where to look for history logs

```python
width20catsExp = ['4', '3', '5', '6', '7']
width20catsLabels = ['384, 1152, 2304', '256, 896, 1792', '192, 640, 1216', '128, 448, 896', '64, 256, 480']

width53catsExp = ['2_4', '2_5' , '2_6', '2_7', '2_8']
width53catsLabels = ['256, 896, 1792', '192, 640, 1216', '128, 448, 896','64, 256, 480', '32, 128, 256']

depth53catsExp = ['3_8', '3_7', '2_7', '3_9']
depth53catsLabels = ['1, 2, 1', '2, 4, 2', '5, 10, 5', '8, 16, 8']

mod53catsExp = ['2_7', '4_7', '4_8', '4_9', '4_10']
mod53catsLabels = ['No modification', 'stem modification', 'inceptionA modification 1', 'inceptionA modification 2', 'inceptionB modification']
```

This is the path that all the saved data regardinng model runs are located in my Google Drive

```python
modelFolder = "/content/drive/MyDrive/Research/Project_Inception_ResNet/RM-Experiments/trainedModelE"
```

A utility method for getting the paths to history files from the abovementioned lists

```python
def getHistoryPaths(expList):
  pathList = []
  for experiment in expList:
    pathList.append(modelFolder + experiment + '/' + 'trainHistory')
  return pathList
```

Plotting utility methods

```python
def accuracyPlotter(expList, labelList, isValidationAccuracy = True):
  #Loading history pickle files:
  historyPathList = getHistoryPaths(expList)
  for path in historyPathList:
    with open(path, 'rb') as file:
      history = pickle.load(file)
      if isValidationAccuracy:
        plt.plot(history['val_accuracy'])
      else:
        plt.plot(history['accuracy'])
  plt.ylabel('accuracy')
  plt.xlabel('epoch')
  plt.legend(labelList, loc='lower right')
  plt.savefig(getFigPath(expList, isValidation = isValidationAccuracy, isAccuracy = True) + '.eps', format='eps')

def lossPlotter(expList, labelList, isValidationLoss = True):
  #Loading history pickle files:
  historyPathList = getHistoryPaths(expList)
  for path in historyPathList:
    with open(path, 'rb') as file:
      history = pickle.load(file)
      if isValidationLoss:
        plt.plot(history['val_loss'])
      else:
        plt.plot(history['loss'])
  plt.ylabel('loss')
  plt.xlabel('epoch')
  plt.legend(labelList, loc='upper right')
  plt.savefig(getFigPath(expList, isValidation = isValidationLoss, isAccuracy = False) + '.eps', format='eps')



def getFigPath(expList, isValidation, isAccuracy):
  dir = "/content/drive/MyDrive/Research/Project_Inception_ResNet/RM-Experiments/plots/"
  filename = ""
  if expList == width20catsExp:
    filename += "width20"
  elif expList == width53catsExp:
    filename += "width53"
  elif expList == depth53catsExp:
    filename += "depth53"
  elif expList == mod53catsExp:
    filename += "mod53"
  if isValidation:
    filename += "val"
  else:
    filename += "train"
  if isAccuracy:
    filename +="_accuracy"
  else:
    filename +="_loss"
  return dir + filename
```
